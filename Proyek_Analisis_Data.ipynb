### Cleaning Data

```python
# 1. Handling Missing Values in Order Reviews
order_reviews_df['review_comment_title'] = order_reviews_df['review_comment_title'].fillna('')
order_reviews_df['review_comment_message'] = order_reviews_df['review_comment_message'].fillna('')

print("\nMissing values in 'review_comment_title' and 'review_comment_message' handled:")
print(order_reviews_df[['review_comment_title', 'review_comment_message']].isnull().sum())

# 2. Handling Missing Values in Orders
orders_cleaned = orders_df.dropna(subset=['order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date'])
print("\nMissing values in 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date' handled:")
print(orders_cleaned.isnull().sum())

# 3. Handling Missing Values in Products
products_df['product_category_name'] = products_df['product_category_name'].fillna('Unknown')
products_df['product_name_lenght'] = products_df['product_name_lenght'].fillna(products_df['product_name_lenght'].median())
products_df['product_description_lenght'] = products_df['product_description_lenght'].fillna(products_df['product_description_lenght'].median())
products_df['product_photos_qty'] = products_df['product_photos_qty'].fillna(products_df['product_photos_qty'].median())

print("\nMissing values in 'product_category_name', 'product_name_lenght', 'product_description_lenght', 'product_photos_qty' handled:")
print(products_df[['product_category_name', 'product_name_lenght', 'product_description_lenght', 'product_photos_qty']].isnull().sum())

# 4. Handling Zero Weights in Products
median_weight = products_df['product_weight_g'].median()
products_df['product_weight_g'] = products_df['product_weight_g'].replace(0, median_weight)

print("\nZero product weights handled (replaced with median weight):")
print(products_df[products_df['product_weight_g'] == 0])

# 5. Removing Duplicates
customers_cleaned = customers_df.drop_duplicates()
geolocations_cleaned = geolocation_df.drop_duplicates()
order_items_cleaned = order_items_df.drop_duplicates()
order_payments_cleaned = order_payments_df.drop_duplicates()
order_reviews_cleaned = order_reviews_df.drop_duplicates()
sellers_cleaned = sellers_df.drop_duplicates()

print("\nDuplicate rows removed from datasets:")
print(f"Customers: {customers_cleaned.shape}")
print(f"Geolocations: {geolocations_cleaned.shape}")
print(f"Order Items: {order_items_cleaned.shape}")
print(f"Order Payments: {order_payments_cleaned.shape}")
print(f"Order Reviews: {order_reviews_cleaned.shape}")
print(f"Sellers: {sellers_cleaned.shape}")

# 6. Ensuring Data Time Consistency
orders_cleaned = orders_cleaned.dropna(subset=['order_delivered_customer_date', 'order_delivered_carrier_date'])
print("\nEnsured data consistency for 'order_delivered_customer_date' and 'order_delivered_carrier_date':")
print(orders_cleaned.isnull().sum())
```

**Insight:**
- Missing values di kolom review_comment_title dan review_comment_message telah diisi dengan string kosong, memastikan bahwa tidak ada missing values yang tersisa dalam kolom ulasan.
- Missing values pada kolom order_approved_at, order_delivered_carrier_date, dan order_delivered_customer_date berhasil ditangani dengan menghapus baris yang memiliki nilai kosong, sehingga data pesanan menjadi lebih bersih dan lengkap.
- Missing values dalam products_df untuk kolom seperti product_category_name, product_name_lenght, product_description_lenght, dan product_photos_qty telah diisi dengan nilai median atau default seperti "Unknown" untuk memastikan data tetap konsisten dan tidak ada missing values.
- Produk dengan berat 0 telah diperbaiki dengan mengganti nilai berat nol dengan nilai median, mengatasi masalah produk dengan data yang tidak valid (berat nol).
- Semua duplikasi data telah dihapus dari berbagai dataset seperti customers_df, geolocation_df, order_items_df, order_payments_df, order_reviews_df, dan sellers_df, memastikan bahwa data unik dan tidak ada redundansi yang dapat mempengaruhi analisis.
- Konsistensi waktu pada kolom order_delivered_customer_date dan order_delivered_carrier_date telah diperiksa dan diperbaiki, memastikan bahwa data pengiriman sesuai dengan urutan waktu yang benar.

### Merge All data

#### Merge orders_df dengan customers_df

```python
# Merge orders_df dengan customers_df berdasarkan 'customer_id'
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.sample(5)
```

#### Merge order_items_df dengan sellers_df

```python
items_sellers_df = pd.merge(
    left=order_items_df,
    right=sellers_df,
    how="left",
    left_on="seller_id",
    right_on="seller_id"
)

items_sellers_df.sample(5)
```

#### Merge products_df dengan products_category_name_df

```python
products_full_df = pd.merge(
    left=products_df,
    right=products_category_name_df,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)

products_full_df.sample(5)
```

####  Merge order_payments_df dengan order_reviews_df

```python
# Merge order_payments_df dengan order_reviews_df berdasarkan 'order_id'
payments_reviews_df = pd.merge(
    left=order_payments_df,
    right=order_reviews_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)

payments_reviews_df.sample(5)
```

#### Merge All Data

```python
# Merge orders_customers_df dengan items_sellers_df berdasarkan 'order_id'
merged_data_1 = pd.merge(
    left=orders_customers_df,
    right=items_sellers_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)

# Merge merged_data_1 dengan products_full_df berdasarkan 'product_id'
merged_data_2 = pd.merge(
    left=merged_data_1,
    right=products_full_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)

# Merge merged_data_2 dengan payments_reviews_df berdasarkan 'order_id'
final_merged_data = pd.merge(
    left=merged_data_2,
    right=payments_reviews_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)

final_merged_data.sample(5)
```

### Pertanyaan 1: Bagaimana tingkat kepuasan customer terhadap layanan kami?

```python
review_scores = final_merged_data['review_score'].value_counts().sort_values(ascending=False)
most_common_score = review_scores.idxmax()

sns.set(style="darkgrid")

plt.figure(figsize=(10, 5))
sns.barplot(
    x=review_scores.index,
    y=review_scores.values,
    order=review_scores.index,
    palette=["green" if score == most_common_score else "grey" for score in review_scores.index],
)

plt.title("Customer Satisfaction Ratings", fontsize=15)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.xticks(fontsize=12)
plt.show()
```
